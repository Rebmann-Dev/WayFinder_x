```python
import pandas as pd
import numpy as np

# load 2023 country data
c2023_path = "../data/country_safety/country_crime_pop_2023_clean.csv"  
c2023 = pd.read_csv(c2023_path)

print("2023 raw columns:", c2023.columns.tolist())
print(c2023.head())
print(c2023.shape)
```

```python
# clean column names
c2023.columns = (
    c2023.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

# normalize country names
c2023["country"] = c2023["country"].astype(str).str.strip()
c2023["country_norm"] = c2023["country"].str.lower()

# ensure numeric types
num_cols_23 = ["crime_index_2023", "population_2023"]
for c in num_cols_23:
    if c in c2023.columns:
        c2023[c] = pd.to_numeric(c2023[c], errors="coerce")

print("2023 describe numeric:")
print(c2023[num_cols_23].describe())
print("2023 missing fractions:")
print(c2023[num_cols_23].isna().mean())

# duplicated country_norm check
dupes_23 = c2023[c2023["country_norm"].duplicated(keep=False)].sort_values("country_norm")
print("2023 duplicated countries (if any):")
print(dupes_23[["country", "country_norm", "crime_index_2023"]].head(20))
print(c2023.shape)
```

```python
# load 2020 country data
c2020_path = "../data/country_safety/country_crime_safety_2020_clean.csv" 
c2020 = pd.read_csv(c2020_path)

print("\n2020 raw columns:", c2020.columns.tolist())
print(c2020.head())
print(c2020.shape)
```

```python
# clean column names
c2020.columns = (
    c2020.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

# normalize country names
c2020["country"] = c2020["country"].astype(str).str.strip()
c2020["country_norm"] = c2020["country"].str.lower()

num_cols_20 = ["crime_index_2020", "safety_index_2020"]
for c in num_cols_20:
    if c in c2020.columns:
        c2020[c] = pd.to_numeric(c2020[c], errors="coerce")

print("2020 describe numeric:")
print(c2020[num_cols_20].describe())
print("2020 missing fractions:")
print(c2020[num_cols_20].isna().mean())

dupes_20 = c2020[c2020["country_norm"].duplicated(keep=False)].sort_values("country_norm")
print("2020 duplicated countries (if any):")
print(dupes_20[["country", "country_norm", "crime_index_2020"]].head(20))
print(c2020.shape)
```

```python
# quick overlap and coverage check
keys_23 = set(c2023["country_norm"])
keys_20 = set(c2020["country_norm"])

print("\nnum countries 2023:", len(keys_23))
print("num countries 2020:", len(keys_20))
print("common countries:", len(keys_23 & keys_20))
print("only in 2023:", len(keys_23 - keys_20))
print("only in 2020:", len(keys_20 - keys_23))

print("\nsample only-2023 countries:", list(keys_23 - keys_20)[:15])
print("sample only-2020 countries:", list(keys_20 - keys_23)[:15])
```
